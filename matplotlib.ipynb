```python
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['font.family'] = 'Hiragino sans'
# plt.rcParams['font.family'] = 'AppleGothic' 
plt.style.use("ggplot")
plt.rcParams['image.cmap'] = 'jet'
```

# MATLAB,オブジェクト指向スタイルの違い

```python
# MATLABスタイルでの描画

x = [1,2,3]
y = [1,4,9]

plt.plot(x,y) # 折線グラフを描画
plt.title('MATLAB-style') # グラフにタイトルを設定

plt.show() # グラフを表示
```

```python
# オブジェクト指向スタイル

x = [1,2,3]
y = [1,4,9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig,ax = plt.subplots()

ax.plot(x,y) # 折れ線グラフを描画
ax.set_title('OOP-style') # タイトルを設定

plt.show() # グラフを表示
```

## オブジェクト指向スタイルの使い方
- オブジェクト指向スタイルは一見冗長
- けれど、1つのfigごとに色々設定できて便利
- 構造としては、１つのfigに対して、１つ以上のsubplotを配置するもの

### subplotってなんだ

```python
# 1つのfigに対して、２つのサブプロットを配置した例

fig, axes = plt.subplots(2)
plt.show()
```

```python
# 1つのfigに対して、２*2のサブプロットを配置した例

fig, axes = plt.subplots(2,2)
plt.show()
```

# グラフの種類
## 折線グラフ(plotメソッド)
- 値の間隔を細かくすれば、擬似的に玉泉のグラフも作成できる

```python
fix ,ax = plt.subplots()

x = [1,2,3]
y1 = [1,2,3]
y2 = [4,1,2]

ax.plot(x,y1)
ax.plot(x,y2)

plt.show()
```

```python
import numpy as np

x = np.arange(0.0 , 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fix ,ax = plt.subplots()

ax.plot(x,y1,label='sin')
ax.plot(x,y2,label='cos')
ax.legend()

plt.show()
```

## 棒グラフ(barメソッド,barhメソッド)

```python
fix ,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

ax.bar(x,y)

plt.show()
```

```python
fix ,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

labels = ['spam', 'ham', 'egg']
ax.bar(x,y,tick_label=labels) # barメソッドにtick_label引数を使用して、目盛りに任意のラベルを作成

plt.show()
```

```python
# 横向き棒グラフはbarhメソッド
fix ,ax = plt.subplots()

x = [1,2,3]
y = [10,2,3]

labels = ['spam', 'ham', 'egg']
ax.barh(x,y,tick_label=labels)

plt.show()
```

```python
# 複数の棒グラフを並べて表示するには...
fix ,ax = plt.subplots()

width = 0.4 # 棒グラフの幅を0.4にする

x1 = [1,2,3]
x2 = [num + width for num in x1] # x1+widthのx軸の値をつくる
y1 = [10,2,3]
y2 = [5,3,6]
labels = ['spam', 'ham', 'egg']


ax.bar(x1,y1,width=width,tick_label=labels,label='y1')
ax.bar(x2,y2,width=width,tick_label=labels,label='y2')

ax.legend()

plt.show()
```

```python
# 積み上げ棒グラフはこうする
# y_totalを描画した後にy2を重ねて描画することで、積上棒グラフに見せてる感じ？
fix ,ax = plt.subplots()

x = [1,2,3]
y1 = [10,2,3]
y2 = [5,3,6]
labels = ['spam', 'ham', 'egg']

# y_totalを計算
y_total = [num1 + num2 for num1,num2 in zip(y1,y2)]


ax.bar(x,y_total,tick_label=labels,label='y1')
ax.bar(x,y2,label='y2')

ax.legend()

plt.show()
```

# ヒストグラム(histメソッド)

```python
np.random.seed(123)
mu = 100 # 平均
sigma = 15 # 標準偏差

# 平均と標準偏差を指定した分布で、乱数を作成する
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
n, bins, patches = ax.hist(x)

plt.show()
```

- ax.hist(x)には３つの情報が格納されている
    - n = ax.hist(x)[0]には各binの度数
    - bins = ax.hist(x)[1]にはbinの境界の値（x軸のどこに描画するかの位置情報的なもの）
    - patches = ax.hist(x)[2]にはbinを描くための情報
        - <BarContainer object of 10 artists> こういうのが入っている

```python
# ビンの数を指定したいとき
fif, ax =plt.subplots()

ax.hist(x, bins=25)

plt.show()
```

```python
# 横向きにする場合
fif, ax =plt.subplots()

ax.hist(x, orientation="horizontal")

plt.show()
```

```python
# 並べて表示したいときは、棒グラフと違って自動で表示される
np.random.seed(123)
mu = 100

# データ生成
x0 = np.random.normal(mu, 20, 1000)
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

# ヒストグラムを描画する一連の操作
fig, ax = plt.subplots()

labels = ['x0','x1','x2']
ax.hist((x0,x1,x2),label=labels) # こうすると並べて表示できる
ax.legend()

plt.show()
```

```python
# 積み上げヒストグラムにするとき
# histメソッドの引数にstacked=Trueを指定

# ヒストグラムを描画する一連の操作
fig, ax = plt.subplots()

labels = ['x0','x1','x2']
ax.hist((x0,x1,x2), stacked=True, label=labels) # こうすると並べて表示できる
ax.legend()

plt.show()
```

# 箱ヒゲ図(boxplotメソッド)

```python
np.random.seed(123)

# データ生成
x0 = np.random.normal(0, 10, 500)
x1 = np.random.normal(0, 15, 500)
x2 = np.random.normal(0, 20, 500)

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']

ax.boxplot((x0,x1,x2), labels=labels)

plt.show()
```

# 円グラフ(pieメソッド)
- デフォルトでは3時の位置から半時計回りで描画される

```python
labels = ['spam', 'ham', 'egg']
x = [10,3,1]

fig, ax = plt.subplots()

ax.pie(x, labels=labels)
ax.axis('equal') # アスペクト比を保持して描画する。≒楕円にならないようにする

plt.show()
```

```python
# 12時の位置から、時計回りに描画してみる
# pieメソッドの引数startangle=90, counterclock=Faleseをそれぞれ指定する

fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False)
ax.axis('equal') 

plt.show()
```

```python
# 値のパーセント表記を追加するには
# autopct='%1.2f%%'を指定する

fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, autopct='%1.2f%%')
ax.axis('equal') 

plt.show()
```

```python
# 一部を目立たせたいとき
# 引数explodeを指定してあげる

explode = [0,0.2,0] # 2つめの要素を20%外にはみ出させる

fig, ax = plt.subplots()

ax.pie(x, labels=labels, startangle=90, counterclock=False, autopct='%1.2f%%', explode=explode)
ax.axis('equal') 

plt.show()
```

# 複数のグラフを組み合わせる

```python
fig, ax = plt.subplots()

x1 = [1,2,3]
y1 = [5,2,3]
x2 = [1,2,3,4]
y2 = [8,5,4,6]

ax.bar(x1,y1,label='y1')
ax.plot(x2,y2,label='y2')
ax.legend()

plt.show()
```

```python
# ヒストグラムと折れ線グラフ
np.random.seed(123)
x = np.random.randn(1000) # 正規乱数を生成

fig, ax = plt.subplots()

# ヒストグラムを描画
counts, edges, patches = ax.hist(x, bins=25)

# 近似曲線に用いる点を求める
x_fit = (edges[:-1] + edges[1:]) / 2
# 近似曲線を描画
y = 1000 * np.diff(edges) * np.exp(-x_fit**2 /2) / np.sqrt(2 * np.pi)
ax.plot(x_fit,y)

plt.show()
```

# スタイルってなんだ
- いろいろ設定できる
    - 色
    - 線
    - フォント
    - テキスト

```python
# 色を変えたいなら、color引数を指定する
fig, ax = plt.subplots()

ax.plot([1,2,3],[1,4,9], label = "色を変えてみた", color='aqua')
ax.legend(loc='best')

plt.show()
```

```python
# 線を変える時
# 線の幅はlinewidthを指定
# 線の種類はlinestyleを指定

fig, ax = plt.subplots()

ax.plot([1,2,3],[1,4,9], linestyle='--', linewidth=10, label = "太い波線にしてみた")
ax.legend(loc='best')

plt.show()
```

```python
# フォントは引数family,size,widthなどで変えられる
# けど、使う気がしないので省略

# 引数fontdictを使うと、まとめて変更できるので便利
```

```python
# テキスト描画
# textメソッドを使用して、グラフないに任意のテキストを描画できる

fig, ax = plt.subplots()
ax.text(0.2,0.4,'テキストです',size=20)

plt.show()
```
